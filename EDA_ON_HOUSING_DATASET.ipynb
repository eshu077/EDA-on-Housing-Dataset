```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Housing.csv')

# Set up the aesthetics for the plots
sns.set_style("whitegrid")

# Create area bins to group houses by their area size
# Bins are created for every 2000 sq. ft.
bins = np.arange(df['area'].min(), df['area'].max() + 2000, 2000)
labels = [f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins)-1)]
df['area_group'] = pd.cut(df['area'], bins=bins, labels=labels, right=False)

# Calculate the average price for each area group, separated by airconditioning
# as_index=False turns the grouped columns back into regular columns
price_trend = df.groupby(['area_group', 'airconditioning'])['price'].mean().reset_index()


# Create the line plot
plt.figure(figsize=(14, 8))
sns.lineplot(
    data=price_trend,
    x='area_group',
    y='price',
    hue='airconditioning', # Creates separate lines for 'yes' and 'no'
    style='airconditioning', # Gives each line a different marker
    markers=True,
    dashes=False,
    linewidth=2.5
)

# Add titles and labels
plt.title('Average House Price Trend by Area and Airconditioning', fontsize=16)
plt.xlabel('Area Group (in sq. ft.)', fontsize=12)
plt.ylabel('Average Price (in millions)', fontsize=12)
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.legend(title='Has Airconditioning?')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Create a figure for multiple subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Scatter plot for Price vs. Area
sns.scatterplot(x='area', y='price', data=df, ax=axes[0])
axes[0].set_title('Price vs. Area', fontsize=14)
axes[0].set_xlabel('Area (sq. ft.)', fontsize=12)
axes[0].set_ylabel('Price (in millions)', fontsize=12)

# Scatter plot for Price vs. Bedrooms
sns.scatterplot(x='bedrooms', y='price', data=df, ax=axes[1])
axes[1].set_title('Price vs. Bedrooms', fontsize=14)
axes[1].set_xlabel('Number of Bedrooms', fontsize=12)
axes[1].set_ylabel('') # Hide y-axis label for clarity

# Scatter plot for Price vs. Bathrooms
sns.scatterplot(x='bathrooms', y='price', data=df, ax=axes[2])
axes[2].set_title('Price vs. Bathrooms', fontsize=14)
axes[2].set_xlabel('Number of Bathrooms', fontsize=12)
axes[2].set_ylabel('') # Hide y-axis label for clarity

plt.tight_layout()
plt.show()
```

```python
# Create a figure for subplots
plt.figure(figsize=(18, 12))

# Subplot 1: Airconditioning
plt.subplot(2, 2, 1)
sns.boxplot(x='airconditioning', y='price', data=df)
plt.title('Price vs. Airconditioning', fontsize=14)
plt.xlabel('Airconditioning', fontsize=12)
plt.ylabel('Price (in millions)', fontsize=12)

# Subplot 2: Furnishing Status
plt.subplot(2, 2, 2)
sns.boxplot(x='furnishingstatus', y='price', data=df)
plt.title('Price vs. Furnishing Status', fontsize=14)
plt.xlabel('Furnishing Status', fontsize=12)
plt.ylabel('')

# Subplot 3: Preferred Area
plt.subplot(2, 2, 3)
sns.boxplot(x='prefarea', y='price', data=df)
plt.title('Price vs. Preferred Area', fontsize=14)
plt.xlabel('Preferred Area', fontsize=12)
plt.ylabel('Price (in millions)', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Create a copy of the dataframe to avoid changing the original
df_numeric = df.copy()

# List of columns to convert from 'yes'/'no' to 1/0
binary_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Convert binary columns to 1s and 0s
for col in binary_cols:
    df_numeric[col] = df_numeric[col].map({'yes': 1, 'no': 0})

# Convert 'furnishingstatus' into dummy variables
df_numeric = pd.get_dummies(df_numeric, columns=['furnishingstatus'], drop_first=True)

# Calculate the correlation matrix
correlation_matrix = df_numeric.corr()

# Create the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Housing Features', fontsize=16)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Housing.csv')

# Set up the aesthetics for the plots
sns.set_style("whitegrid")

# Create a smaller dataframe for the pairplot for clarity
df_pairplot = df[['price', 'area', 'bedrooms', 'bathrooms', 'airconditioning']]

# Create the pairplot
sns.pairplot(df_pairplot, hue='airconditioning', plot_kws={'alpha': 0.6})
plt.suptitle('Pair Plot of Numerical Features by Airconditioning', y=1.02, fontsize=16)
plt.show()
```

```python
# Create a figure for subplots
plt.figure(figsize=(16, 7))

# Subplot 1: Guest Room
plt.subplot(1, 2, 1)
sns.violinplot(x='guestroom', y='price', data=df)
plt.title('Price Distribution by Guest Room', fontsize=14)
plt.xlabel('Guest Room', fontsize=12)
plt.ylabel('Price (in millions)', fontsize=12)

# Subplot 2: Basement
plt.subplot(1, 2, 2)
sns.violinplot(x='basement', y='price', data=df)
plt.title('Price Distribution by Basement', fontsize=14)
plt.xlabel('Basement', fontsize=12)
plt.ylabel('') # Hide y-axis label for clarity

plt.show()
```

```python
# Create a figure for subplots
plt.figure(figsize=(20, 5))

# Subplot 1: Main Road
plt.subplot(1, 4, 1)
sns.countplot(x='mainroad', data=df)
plt.title('Count of Houses on Main Road')

# Subplot 2: Hot Water Heating
plt.subplot(1, 4, 2)
sns.countplot(x='hotwaterheating', data=df)
plt.title('Count of Houses with Hot Water')

# Subplot 3: Preferred Area
plt.subplot(1, 4, 3)
sns.countplot(x='prefarea', data=df)
plt.title('Count of Houses in Preferred Area')

# Subplot 4: Furnishing Status
plt.subplot(1, 4, 4)
sns.countplot(x='furnishingstatus', data=df)
plt.title('Count by Furnishing Status')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

sns.scatterplot(x='area', y='price', hue='furnishingstatus', data=df, palette='viridis', s=80)

plt.title('Price vs. Area by Furnishing Status', fontsize=16)
plt.xlabel('Area (sq. ft.)', fontsize=12)
plt.ylabel('Price (in millions)', fontsize=12)
plt.legend(title='Furnishing Status')

plt.show()
```
